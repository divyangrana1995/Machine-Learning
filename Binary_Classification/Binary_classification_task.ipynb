# Data Preprocessing

```python
# Convert the categorigal data to numerical using one hot encoding method
# drop 'ZIK' because over 2000 missing values
cat_dummies = pd.get_dummies(categorical.drop('ZIK', axis=1), drop_first = True)

df= pd.concat([numerical.drop('ID', axis=1), cat_dummies], axis = 1)
df = df.drop(df.filter(regex='ERZ|HUI|POUG|CDx|LPI').columns, axis=1)
print(f'column names:{df.columns}')
print(f'shape:{df.shape}')
df.head(5)
```

```python
# Empty values in each column
df.isnull().sum()
```

```python
#Target variable
Y= df['Type_y']
Y
```

```python
#Input features
X= df.drop('Type_y', axis=1)
X
```

```python
#Standardization using MinMax Scaler
X = pd.DataFrame(scaler.fit_transform(X), columns = X.columns)
X
```

```python
# Correlation matrix checking for new correlations
complete_correlation= associations(df, filename= 'complete_correlation_after_encoding.png', figsize=(30,30))
df_complete_corr=complete_correlation['corr']
```

```python
X= X.drop(['BJKG_hh','GJAH_rrr'], axis=1)
# correlation between 'BJKG_hh' and 'BJZHD_hh' (coeff = 0.82); and between 'GJAH_rrr' and 'GJAH_oooo' (coeff = -0.91)
```

```python
print(X.shape)
```

```python
# Dealing with missing values using KNN imputer
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors = 3)
X = pd.DataFrame(imputer.fit_transform(X),columns = X.columns)
X
```

# Machine Learning models using my own selected features
It is best to check the accuracy of different classifier and then decide the best ot of the lot. Here, the features used to train the classifier are selected after analysing the data.

Support Vector Machine classifier

```python
from sklearn.metrics import f1_score,confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score,classification_report
from sklearn.model_selection import train_test_split

from sklearn import svm

# split data train 70 % and test 30 %
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3,stratify = y, random_state=42)

clf_svm = svm.SVC()
clf_svm.fit(x_train, y_train)

ac = accuracy_score(y_test,clf_svm.predict(x_test))
print('Accuracy is: ',ac)
cm = confusion_matrix(y_test,clf_svm.predict(x_test))
sn.heatmap(cm,annot=True,fmt="d")
print(classification_report(y_test,clf_svm.predict(x_test)))
```

K- Nearest neighbors classifier

```python
from sklearn import neighbors
clf_knn = neighbors.KNeighborsClassifier(3, weights='distance')
clf_knn.fit(x_train, y_train)
ac = accuracy_score(y_test,clf_knn.predict(x_test))
print('Accuracy is: ',ac)
cm = confusion_matrix(y_test,clf_knn.predict(x_test))
sn.heatmap(cm,annot=True,fmt="d")
print(classification_report(y_test,clf_knn.predict(x_test)))
```

Decision Tree classifier

```python
from sklearn import tree
clf_dt= tree.DecisionTreeClassifier()
clf_dt = clf_dt.fit(x_train, y_train)

ac = accuracy_score(y_test,clf_dt.predict(x_test))
print('Accuracy is: ',ac)
cm = confusion_matrix(y_test,clf_dt.predict(x_test))
sn.heatmap(cm,annot=True,fmt="d")
print(classification_report(y_test,clf_dt.predict(x_test)))
```

Random Forest classifier

```python
from sklearn.ensemble import RandomForestClassifier

clf_rf = RandomForestClassifier(random_state=43)      
clr_rf = clf_rf.fit(x_train,y_train)

ac = accuracy_score(y_test,clf_rf.predict(x_test))
print('Accuracy is: ',ac)
cm = confusion_matrix(y_test,clf_rf.predict(x_test))
sn.heatmap(cm,annot=True,fmt="d")
print(classification_report(y_test,clf_rf.predict(x_test)))
```

#  Machine Learning models using Feature selection process in sklearn.
Here, the features that would be used to train the classifier is selected by an in-built sklearn module that selects the best of all the features.

```python
# New dataframe 
df_1= pd.concat([numerical.drop('ID', axis=1), cat_dummies], axis = 1)
X_1= df_1.drop('Type_y', axis=1)
X_1= pd.DataFrame(scaler.fit_transform(X_1), columns = X_1.columns)
X_1 = pd.DataFrame(imputer.fit_transform(X_1),columns = X_1.columns)
# split data train 70 % and test 30 %
x_train_1, x_test_1, y_train_1, y_test_1 = train_test_split(X_1, Y, test_size=0.3,stratify=y, random_state=42)

```

```python
# Selection of features usinf SelectKBest
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
# find best scored k features

select_feature = SelectKBest(chi2, k=15).fit(x_train_1, y_train_1)
```

```python
print('Score list:', select_feature.scores_)
print('Feature list:', x_train_1.columns)
```

```python
# Transform the train and test set of features to selected features by SelectKBest
x_train_2 = select_feature.transform(x_train_1)
x_test_2 = select_feature.transform(x_test_1)
```

Support Vector Machine classifier

```python
clf_svm_2 = svm.SVC()
clf_svm_2.fit(x_train_2, y_train_1)

ac = accuracy_score(y_test_1,clf_svm_2.predict(x_test_2))
print('Accuracy is: ',ac)
cm = confusion_matrix(y_test_1,clf_svm_2.predict(x_test_2))
sn.heatmap(cm,annot=True,fmt="d")
print(classification_report(y_test_1,clf_svm_2.predict(x_test_2)))
```

K nearest neighbor classifier

```python
clf_knn_2 = neighbors.KNeighborsClassifier(3, weights='distance')
clf_knn_2=clf_knn_2.fit(x_train_2, y_train_1)
ac = accuracy_score(y_test_1,clf_knn_2.predict(x_test_2))
print('Accuracy is: ',ac)
cm = confusion_matrix(y_test_1,clf_knn_2.predict(x_test_2))
sn.heatmap(cm,annot=True,fmt="d")
print(classification_report(y_test_1,clf_knn_2.predict(x_test_2)))
```

Decision Tree Classifier

```python
clf_dt_2= tree.DecisionTreeClassifier()
clf_dt_2 = clf_dt_2.fit(x_train_2, y_train_1)

ac = accuracy_score(y_test_1,clf_dt_2.predict(x_test_2))
print('Accuracy is: ',ac)
cm = confusion_matrix(y_test_1,clf_dt_2.predict(x_test_2))
sn.heatmap(cm,annot=True,fmt="d")
print(classification_report(y_test_1,clf_dt_2.predict(x_test_2)))
```

Random Forest Classifier

```python
clf_rf_2 = RandomForestClassifier(random_state=43)      
clr_rf_2 = clf_rf_2.fit(x_train_2,y_train_1)
ac_2 = accuracy_score(y_test_1,clf_rf_2.predict(x_test_2))
print('Accuracy is: ',ac_2)
cm_2 = confusion_matrix(y_test_1,clf_rf_2.predict(x_test_2))
sn.heatmap(cm_2,annot=True,fmt="d")
print(classification_report(y_test_1,clf_rf_2.predict(x_test_2)))
```
